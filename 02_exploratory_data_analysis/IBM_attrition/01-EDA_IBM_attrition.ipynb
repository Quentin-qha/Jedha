6. How are the ages distributed?

7. Is the distribution of ages different for men and women?
Find two ways of investigating the difference

```python
df.groupby("Gender")["Age"].describe()
```

```python
fig = px.histogram(df, x="Age", color="Gender")
fig.show()
```

```python
fig = px.box(df, x = "Gender", y = "Age")
fig.show()
```

```python
df_man = df[df["Gender"] == "Male"]["Age"]
df_woman = df[df["Gender"] == "Female"]["Age"]

df_man_math = pd.DataFrame(df_man.describe())
df_woman_math = pd.DataFrame(df_woman.describe())
df_man_math.columns = ["Male"]
df_woman_math.columns = ["Female"]

df_all_gender = df_man_math.merge(df_woman_math, left_index=True, right_index=True) 
df_all_gender["Male"] = df_all_gender["Male"].round(2)
df_all_gender["Female"] = df_all_gender["Female"].round(2)
df_all_gender
```

19. Now we would like to investigate the relationship between attrition and education field. We can't proceed like the previous analyses, as attrition and education field are both categorical variables! Let's do something else:
* Convert the Attrition column such that "Yes" is mapped to 1 and "No" is mapped to 0
* Then, compute the average of Attrition grouped by the education field
* Finally, plot the average attrition as a function of the education field.
* What is the education field with the highest attrition rate?

```python
df["Attrition_nbr"] = df["Attrition"].str.strip().str.lower().apply(lambda x: 1 if x == "yes" else 0)
df
df_attr_education_nbr = df.groupby("EducationField")["Attrition_nbr"].mean().sort_values(ascending=False).reset_index()
df_attr_education_nbr
```

```python
fig = px.bar(df_attr_education_nbr, x="EducationField", y="Attrition_nbr")
fig.show()
```
